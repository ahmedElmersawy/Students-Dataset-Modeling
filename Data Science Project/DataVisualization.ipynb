```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd
import geoplot as gplt
data = pd.read_csv("Students_data.csv")
```

#### Bar plot for Students Class Distrubtion

```python
class_counts = data['class'].value_counts()
plt.bar(class_counts.index, class_counts.values)
plt.xlabel('Class')
plt.ylabel('Student Numbers')
plt.title('Class Distribution')
plt.show()
```

```python
gender_counts = data["gender"].value_counts()
plt.pie(gender_counts, labels=gender_counts.index, autopct='%.1f%%')
plt.title('Gender Distribution')
plt.show()
```

```python
plt.scatter(data["ID"], data["GPA"])
plt.xlabel("ID")
plt.ylabel("GPA")
plt.title(" GPA Distribution")
plt.show()
```

```python

plt.hist(data["Algebra"],bins=100 , edgecolor='blue')
plt.xlabel("Algebra")
plt.ylabel("Frequency")
plt.title("Algebra grades")
plt.grid(True)  
plt.show()
```
